### A/B testing and Hypothesis Supershop Data


```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
#load supershop data
df = pd.read_csv('/content/drive/MyDrive/Data Analysis and Machine Learning with python Project/A_B_testing_Supershop_data/supershop_data.csv')
```

### A/B testing

A/B Testing is used to compare two versions (A and B) of a digital product (e.g. website, app feature) to see which one performs better Here,
- We simulate a

 - Version A (Control) – the original

  - Version B (Variant) – a modified version with one change


##Step-1:Summary Statistics for each *group*

```python
df.groupby('Group')['MonthlySpend'].describe()
```

##Step-2: Visualize two groups

```python
sns.boxplot(x='Group', y='MonthlySpend', data=df)
plt.title('Monthly Spend by Group')
plt.show()
```

## Step 3: Hypothesis Testing - Independent T-test

```python
control = df[df['Group'] == 'Control']['MonthlySpend']
variant = df[df['Group'] == 'Variant']['MonthlySpend']
t_stat, p_value = stats.ttest_ind(control, variant)
print(f'T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}')
```

**Interpretation:**
If p-value < 0.05, we reject the null hypothesis, indicating the variant caused a statistically significant change in Monthly Spend
